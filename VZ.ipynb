```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df =  pd.read_csv('adult (1).csv')
df
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=10, color='skyblue')
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.stripplot(data=df,x='education.num', y='age',jitter=True)
plt.title('Dot Plot')
plt.ylabel('Values')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='education.num',color='lightgreen')
plt.title('Bar Plot')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

```python
sns.lineplot(data=df, x='age', y='education.num')
plt.title('Line Chart of Age vs Hours-per-week')
plt.xlabel('Age')
plt.ylabel('Hours-per-week')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='education.num', data=df)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age vs. Cholesterol')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
le = LabelEncoder()
df['income']=le.fit_transform(df['income'])
```

```python
salary_counts = df['income'].value_counts()
plt.pie(salary_counts, labels=['<=50K', '>50K'])
plt.title('Pie chart of Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='age', color='lightblue')
sns.histplot(df['age'], bins=5, kde=False, color='skyblue')
sns.stripplot(data=df, y='age', jitter=True, color='orange')
plt.title('Box Plot with Histogram and Scatter')
plt.ylabel('Values')
plt.show()
```
